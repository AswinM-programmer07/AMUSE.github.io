# **From the Given Equation simple**


```python
#Hodgkin-Huxley Model Equations:
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
#All the input values are extracted from the book "Neuroscience: Exploring the Brain"
# Constants
Cm = 1.0  # Membrane capacitance
# External current stimulation
I_ext = 10
# Sodium (Na+) current  and Nernst potential
g_Na = 120
E_Na = 50
# Potassium (K+) current  and Nernst potential (mV)
g_K = 36
E_K = -77
# Leakage current and Nernst potential (mV)
g_L = 0.3
E_L = -54.4
# Time parameters
dt = 0.01 # Time step (ms)
num_steps = 100  # Number of time steps
time = np.arange(0, num_steps * dt, dt)
# Initialize membrane potential array
V = np.zeros(num_steps)
# Initial membrane potential (mV)
V[0] = -65
# Simulation loop
for i in range(1, num_steps):
    # Calculate ion currents
    I_Na = g_Na * (V[i-1] - E_Na)
    I_K = g_K * (V[i-1] - E_K)
    I_L = g_L * (V[i-1] - E_L)

    # Update membrane potential using the membrane potential equation
    dV_dt = (I_ext - I_Na - I_K - I_L) / Cm
    V[i] = V[i-1] + dV_dt * dt
# Plot membrane potential over time
plt.plot(time, V)
plt.title('Membrane Potential over Time')
plt.xlabel('Time (ms)')
plt.ylabel('Membrane Potential (mV)')
plt.grid(True)
plt.show()
```

# **Ioni current dynamics:**

# All the input values are extracted from the book"**Principles of Computational Modelling in Neuroscience"**

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
V = np.linspace(-100, 100, 1000)  # Membrane potential (mV)

# Rate functions for m, n, and h
def alpha_m(v):
    return 0.1 * (v + 40) / (1 - np.exp(-(v + 40) / 10))

def beta_m(v):
    return 4.0 * np.exp(-(v + 65) / 18)

def alpha_n(v):
    return 0.01 * (v + 55) / (1 - np.exp(-(v + 55) / 10))

def beta_n(v):
    return 0.125 * np.exp(-(v + 65) / 80)

def alpha_h(v):
    return 0.07 * np.exp(-(v + 65) / 20)

def beta_h(v):
    return 1 / (1 + np.exp(-(v + 35) / 10))

# Gating variable dynamics
def dm_dt(v, m):
    return alpha_m(v) * (1 - m) - beta_m(v) * m

def dn_dt(v, n):
    return alpha_n(v) * (1 - n) - beta_n(v) * n

def dh_dt(v, h):
    return alpha_h(v) * (1 - h) - beta_h(v) * h

# Initial conditions
m0 = 0.05  # Initial value of m
n0 = 0.3177  # Initial value of n
h0 = 0.6  # Initial value of h

# Time vector
t = np.linspace(0, 10, 1000)

# Solve the differential equations
m_values = np.zeros_like(t)
n_values = np.zeros_like(t)
h_values = np.zeros_like(t)
m_values[0] = m0
n_values[0] = n0
h_values[0] = h0
dt = t[1] - t[0]
for i in range(1, len(t)):
    m_values[i] = m_values[i-1] + dm_dt(V[i], m_values[i-1]) * dt
    n_values[i] = n_values[i-1] + dn_dt(V[i], n_values[i-1]) * dt
    h_values[i] = h_values[i-1] + dh_dt(V[i], h_values[i-1]) * dt

# Plot the dynamics of m, n, and h
plt.plot(V, m_values, label='m')
plt.plot(V, n_values, label='n')
plt.plot(V, h_values, label='h')
plt.xlabel('Membrane Potential (mV)')
plt.ylabel('Value')
plt.title('Gating Variable Dynamics')
plt.legend()
plt.grid(True)
plt.show()
```

# **Calcium Dynamics and Neurotransmitter Release:**
image.png

# All the input values are extracted from the book"**Calcium: The Molecular Basis of Calcium Action in Biology and Medicine**"

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
k_ca = 0.1  # Decay rate constant
I_ca = 1.0  # Calcium influx

# Time parameters
dt = 0.01  # Time step
t_start = 0
t_end = 100
num_steps = int((t_end - t_start) / dt)
time = np.linspace(t_start, t_end, num_steps)

# Initial calcium concentration
Ca2_plus = np.zeros(num_steps)
Ca2_plus[0] = 1.0  # Initial calcium concentration

# Euler method to solve the differential equation
for i in range(1, num_steps):
    dCa_dt = -k_ca * Ca2_plus[i-1] + I_ca
    Ca2_plus[i] = Ca2_plus[i-1] + dCa_dt * dt

# Plot calcium concentration over time
plt.plot(time, Ca2_plus)
plt.xlabel('Time')
plt.ylabel('[Ca2+]')
plt.title('Calcium Concentration over Time')
plt.grid(True)
plt.show()
```
