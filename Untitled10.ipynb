```python
#GCcontent function
import random
Nucleotides = ["A","C","G","T"]
DNA_ReverseComplement = {'A':'T','T':'A','G':'C','C':'G'}
#checking the sequence to make sure it is a DNA String
def validateSeq(dna_seq):
  tmpseq = dna_seq.upper()
  for nuc in tmpseq:
    if nuc not in Nucleotides:
      return False
  return tmpseq

randDNAStr = ''.join([random.choice(Nucleotides) for nuc in range(20)])
rnd_seq= "TCTTGAAAAAGAAATAAGAAATGTATAA"
validateSeq(rnd_seq)

#calculating nucleotide frequency
rnd_seq= "TCTTGAAAAAGAAATAAGAAATGTATAA"
validateSeq(rnd_seq)

def countNucFrequency(seq):
  tmpFreqDict = {"A":0, "C": 0, "G":0, "T":0}
  for nuc in seq:
    tmpFreqDict[nuc] +=1
  return tmpFreqDict

countNucFrequency(rnd_seq)
#Transcription
def transcription(seq):
  #DNA ->RNA Transcription
  return seq.replace("T","U")

DNA_ReverseComplement = {'A':'T','T':'A','G':'C','C':'G'}
#REVERSE COMPLEMENT
def reverse_complement(seq):
  return ''.join([DNA_ReverseComplement[nuc] for nuc in seq])[::-1]
reverse_complement(rnd_seq)
#gc content function
def gc_content(seq):
  return round((seq.count('C')+seq.count('G'))/len(seq)*100)
gc_content(rnd_seq)

def gc_content_subsec(seq):
  return round((seq.count)('C') + seq.count('G')/ len(seq)*100)

def gc_content_subsec(seq, k = 20):
  res=[]
  for i in range(0, len(seq)- k + 1,k):
    subseq=seq[i:i + k]
    res.append(gc_content(subseq))
  return res
print("gc_content_subsec",gc_content_subsec(rnd_seq, k = 4))
```

```python

DNA_Codons = {
    # 'M' - START, '_' - STOP
    "GCT": "A", "GCC": "A", "GCA": "A", "GCG": "A",
    "TGT": "C", "TGC": "C",
    "GAT": "D", "GAC": "D",
    "GAA": "E", "GAG": "E",
    "TTT": "F", "TTC": "F",
    "GGT": "G", "GGC": "G", "GGA": "G", "GGG": "G",
    "CAT": "H", "CAC": "H",
    "ATA": "I", "ATT": "I", "ATC": "I",
    "AAA": "K", "AAG": "K",
    "TTA": "L", "TTG": "L", "CTT": "L", "CTC": "L", "CTA": "L", "CTG": "L",
    "ATG": "M",
    "AAT": "N", "AAC": "N",
    "CCT": "P", "CCC": "P", "CCA": "P", "CCG": "P",
    "CAA": "Q", "CAG": "Q",
    "CGT": "R", "CGC": "R", "CGA": "R", "CGG": "R", "AGA": "R", "AGG": "R",
    "TCT": "S", "TCC": "S", "TCA": "S", "TCG": "S", "AGT": "S", "AGC": "S",
    "ACT": "T", "ACC": "T", "ACA": "T", "ACG": "T",
    "GTT": "V", "GTC": "V", "GTA": "V", "GTG": "V",
    "TGG": "W",
    "TAT": "Y", "TAC": "Y",
    "TAA": "_", "TAG": "_", "TGA": "_"
}
rnd_seq= "TCTTGAAAAAGAAATAAGAAATGTATAA"
def translate_seq(seq, init_pos=0):
  return [DNA_Codons[seq[pos:pos + 3]] for pos in range(init_pos, len(seq) - 2, 3)]

print("sequence of amino acids",translate_seq(rnd_seq,0))
```
